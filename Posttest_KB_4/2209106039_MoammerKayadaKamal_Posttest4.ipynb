### Normalisasi

```python
yada_2 = yada.copy()

scale =MinMaxScaler()
scaledData= scale.fit_transform(yada_2[['dtir1']])
yada_2['dtir1'] = scaledData

print('kolom dtir1 sebelum Normalisasi')
print(yada['dtir1'].describe())
print('\n')
print('kolom dtir1 setelah Normalisasi')
print(yada_2['dtir1'].describe())
```

### Standarisasi

```python
yada_3 =yada.copy()

scaler = StandardScaler()
scaledData = scaler.fit_transform(yada_3[['dtir1']])
yada_3['dtir1'] = scaledData

print('Standar deviasi pada kolom dtir1 sebelum di standarisasi')
print(np.std(yada['dtir1']))
print('\n')
print('Standar deviasi pada kolom dtir1 setelah di standarisasi')
print(np.std(yada_3['dtir1']))
```

### Lakukan Data cleaning pada data dengan nilai null (jika tidak ada nilai null pada dataset, maka buatlah menjadi ada). Ganti nilai null sesuai ketentuan. (bilangan bulat : median/modus, bilangan desimal : mean, tulisan : modus),untuk itu di sini hanya mempunyai 2 tipe yaitu float dan object.

```python
yada_4= yada.copy()
yada_4['loan_limit'].fillna(yada_4['loan_limit'].mode()[0],inplace=True)
yada_4['approv_in_adv'].fillna(yada_4['approv_in_adv'].mode()[0],inplace=True)
yada_4['loan_purpose'].fillna(yada_4['loan_purpose'].mode()[0],inplace=True)
yada_4['Neg_ammortization'].fillna(yada_4['Neg_ammortization'].mode()[0],inplace=True)
yada_4['income'].fillna(yada_4['income'].mode()[0],inplace=True)
yada_4['dtir1'].fillna(yada_4['dtir1'].mode()[0],inplace=True)

yada_4['rate_of_interest'].fillna(yada_4['rate_of_interest'].mean(),inplace=True)
yada_4['Interest_rate_spread'].fillna(yada_4['Interest_rate_spread'].mean(),inplace=True)
yada_4['Upfront_charges'].fillna(yada_4['Upfront_charges'].mean(),inplace=True)
yada_4['term'].fillna(yada_4['term'].mean(),inplace=True)
yada_4['property_value'].fillna(yada_4['property_value'].mean(),inplace=True)

```
